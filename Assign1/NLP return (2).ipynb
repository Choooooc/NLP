```python
import numpy as np
import pandas as pd
import pickle
from IPython.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
```

```python
stock = pd.read_csv('/Users/gefan/Desktop/stock_data.csv')
index = pd.read_csv('/Users/gefan/Desktop/index_data.csv')
```

```python
merged = stock.merge(index, left_on=['datadate'], right_on=['caldt'], how='left')
merged['filing_date']='NA'
```

```python
merged['datadate'] = pd.to_datetime(merged['datadate'])
merged['datadate'] = merged['datadate'].dt.strftime("%Y-%m-%d")
```

```python
with open('/Users/gefan/Downloads/ticker_date','rb') as handle:
    filings_by_td = pickle.load(handle)
```

```python
for key in filings_by_td.keys():
    for date in filings_by_td[key]:
        if key in merged['tic'].values:
            index = merged[merged['tic'] == key].loc[merged['datadate']==date].index
            merged.loc[index-1,'filing_date']='pre filing date'
            merged.loc[index,'filing_date']='filing date 0'
            merged.loc[index+1,'filing_date']='filing date 1'
            merged.loc[index+2,'filing_date']='filing date 2'
            merged.loc[index+3,'filing_date']='filing date 3'      
```

```python
merged['year']=merged['datadate'].str[0:4]
merged['td_count']= 0 
```

```python
for tic in merged['tic'].unique():
    for year in merged['year'].unique():
        if year in list(merged[merged['tic']==tic]['year'].unique()):
            if 'filing date 0' in list(merged[(merged['tic']==tic)&(merged['year']==year)]['filing_date'].unique()):
                k_date = max(merged[(merged['tic'] == tic)&(merged['year'] ==  year)&(merged['filing_date'] == 'filing date 0')]['datadate'])
                td_count= len(merged[(merged['tic'] == tic)&(merged['year'] ==  year)&(merged['datadate'] <k_date)])
                index = list(merged[(merged['tic'] == tic)&(merged['year'] ==  year)].index)
                merged.loc[index,'td_count']=td_count
```

```python
pd.set_option('display.max_rows', None)
merged.head(1000)
```

```python
df= merged[merged['filing_date']!='NA']
```

```python
#1 prefiling price >3
low = list(merged[merged['filing_date'] == 'pre filing date'].loc[merged['prccd']<=3].index)
for i in low:
    df = df.drop([i, i+1, i+2, i+3,i+4])
```

```python
#2 60 trading days
for tic in df['tic'].unique():
    if len(df[df['tic']==tic]['year'].unique())<2:
        index = list(df[df['tic']==tic].index)
        df.loc[index,'td_count']=0
    else:
        year_list = sorted(list(df[df['tic']==tic]['year'].unique()))
        for i in range(0,len(year_list)-1):
            if list(df[(df['tic']==tic)&(df['year']==year_list[i])]['td_count'])[0]<60 or list(df[(df['tic']==tic)&(df['year']==year_list[i+1])]['td_count'])[0]<60:
                index1 = list(df[(df['tic'] == tic)&(df['year'] == year_list[i])].index)
                index2 = list(df[(df['tic'] == tic)&(df['year'] == year_list[i+1])].index)
                df.loc[index1,'td_count']=0
                df.loc[index2,'td_count']=0
```

```python
df=df[df['td_count']>=60]
```

```python
df.head(1000)
```

```python
df['excess_return']=0
index = list(df[df['filing_date'] == 'filing date 3'].index)
for i in index:
    stock_return = df.loc[i,'prccd']/df.loc[i-4,'prccd']-1
    index_return = (1+df.loc[i-3,'vwretx'])*(1+df.loc[i-2,'vwretx'])*(1+df.loc[i-1,'vwretx'])*(1+df.loc[i,'vwretx'])-1
    excess_return = stock_return - index_return
    df.loc[i-3,'excess_return']=excess_return
```
